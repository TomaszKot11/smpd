```python
import math
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# calculates distance between two points
def getDistance(x1,y1,x2,y2) :  
    distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)  
    return distance

# checks if point includes x and y values 
def checkIfNotNan(x,y) :
    if(math.isnan(x) and math.isnan(y)) :
        return False
    else :
        return True

try:
    df = pd.read_csv('C://Users//monik//Desktop//leaf//leaf.csv', header=None)
    data = df[(df[0]==3)|(df[0]==5)]
    train, test = train_test_split(data.copy(), test_size=0.2, random_state=222)

    plt.scatter(x=train[5][train[0]==3], y=train[7][train[0]==3], c="red")
    plt.scatter(x=train[5][train[0]==5], y=train[7][train[0]==5], c="blue")

    # train values from class 3
    X_train_3 = train[5][train[0]==3].values
    Y_train_3 = train[7][train[0]==3].values
    
    # train values from class 5
    X_train_5 = train[5][train[0]==5].values
    Y_train_5 = train[7][train[0]==5].values
    
    #test values
    X_test = test[5].values
    Y_test = test[7].values
    
    #result DataFrame object with x value, y value and class classified for point
    results = pd.DataFrame(columns=['x_value', 'y_value', 'class'])

    for i in range(len(X_test)) :
        # ancillary variable containing all distances between test point and points from class 3 and 5
        distances = pd.DataFrame(columns=['distance', 'class'])
        if(checkIfNotNan(X_test[i], Y_test[i])) :
            for j in range(len(X_train_3)) :
                if(checkIfNotNan(X_train_3[i], X_train_3[i])) :
                    distances = distances.append({'distance': getDistance(X_test[i], Y_test[i], X_train_3[j], X_train_3[j]),'class': '3'}, ignore_index=True)
            for k in range(len(X_train_5)) :
                if(checkIfNotNan(X_train_5[i], Y_train_5[i])) :
                    distances = distances.append({'distance': getDistance(X_test[i], Y_test[i], X_train_5[k], Y_train_5[k]),'class': '5'}, ignore_index=True)
            distancesArray = distances['distance'].values
            index = np.argmin(distancesArray)
            results = results.append({'x_value': X_test[i], 'y_value': Y_test[i], 'class': distances.iloc[index,:]['class']}, ignore_index=True)

    print(results)

except IOError :
    print("File not accessible")
```
